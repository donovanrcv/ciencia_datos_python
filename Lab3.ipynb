### 1) ¿Cuál es el tamaño en bytes de un array de 1000 elementos de tipo booleano?

```python
# Crear un array de 1000 elementos de tipo booleano
array = np.ones(1000, dtype=bool)

# Obtener el tamaño en bytes de un elemento del array
tamaño_elemento_bytes = array.itemsize

# Calcular el tamaño total del array en bytes
tamaño_total_bytes = len(array) * tamaño_elemento_bytes

print("El tamaño en bytes de un array de 1000 elementos de tipo booleano es:", tamaño_total_bytes)
```

### 4) Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], obtenga la media aritmética de la diferencia (a-b).

```python
# Definir los arrays
a = [1, 4, 2, 7]
b = [1, 3, 2, 9]

# Calcular la diferencia entre los elementos de los arrays
diferencia = [a[i] - b[i] for i in range(len(a))]

# Calcular la media aritmética de la diferencia
media_diferencia = sum(diferencia) / len(diferencia)

print("La media aritmética de la diferencia (a-b) es:", media_diferencia)
```

### 2) Realice una gráfica de barras sobre la población de cada ciudad, ordénela de menor a mayor y responda: a. ¿Cuál es la segunda ciudad más poblada? b. ¿Qué posición ocupa Granada en el ranking de las ciudades más pobladas?

```python
# Ordenar el dataframe por población de menor a mayor
data_sorted = dataset.sort_values(by='Poblacion')

# Crear la gráfica de barras
plt.figure(figsize=(10, 6))
plt.barh(data_sorted['Municipio'], data_sorted['Poblacion'])
plt.xlabel('Población')
plt.ylabel('Municipio')
plt.title('Población de cada municipio')
plt.grid(axis='x')

# Mostrar la gráfica
plt.show()

# Responder a las preguntas
segunda_ciudad_mas_poblada = data_sorted.iloc[-2]['Municipio']
posicion_granada = data_sorted[data_sorted['Municipio'] == 'Granada'].index[0] + 1

print("a. La segunda ciudad más poblada es:", segunda_ciudad_mas_poblada)
print("b. Granada ocupa la posición número:", posicion_granada, "en el ranking de las ciudades más pobladas.")
```

```python
# Encontrar el municipio situado más al norte
municipio_mas_al_norte = dataset.loc[dataset['Y'].idxmax()]

nombre_municipio_mas_al_norte = municipio_mas_al_norte['Municipio']
provincia_municipio_mas_al_norte = municipio_mas_al_norte['Provincia']
poblacion_municipio_mas_al_norte = municipio_mas_al_norte['Poblacion']

print("El municipio situado más al norte es:", nombre_municipio_mas_al_norte)
print("Pertenece a la provincia de:", provincia_municipio_mas_al_norte)
print("Tiene una población de:", poblacion_municipio_mas_al_norte, "habitantes.")
```

### 5) Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.

```python
# Filtrar los municipios de la provincia de Granada
municipios_granada = dataset[dataset['Provincia'] == 'Granada']

# Calcular las estadísticas sobre la población
media_poblacion = municipios_granada['Poblacion'].mean()
mediana_poblacion = municipios_granada['Poblacion'].median()
desviacion_estandar_poblacion = municipios_granada['Poblacion'].std()
maximo_poblacion = municipios_granada['Poblacion'].max()
minimo_poblacion = municipios_granada['Poblacion'].min()

print("Estadísticas de la población de los municipios de la provincia de Granada:")
print("Media:", media_poblacion)
print("Mediana:", mediana_poblacion)
print("Desviación estándar:", desviacion_estandar_poblacion)
print("Valor máximo:", maximo_poblacion)
print("Valor mínimo:", minimo_poblacion)
```

### 6) Realice un histograma con la población de los Municipios para cada una de las provincias.

```python
# Crear una figura para los histogramas
plt.figure(figsize=(12, 8))

# Iterar sobre cada provincia
for provincia, group_data in dataset.groupby('Provincia'):
    # Trazar un histograma para la población de los municipios de la provincia actual
    plt.hist(group_data['Poblacion'], bins=20, alpha=0.7, label=provincia)

# Configurar el título y etiquetas de los ejes
plt.title('Histograma de Población de Municipios por Provincia')
plt.xlabel('Población')
plt.ylabel('Frecuencia')

# Agregar una leyenda
plt.legend()

# Mostrar el histograma
plt.show()
```

### 7) Seleccione al azar cincuenta municipios diferentes de entre los disponibles en el archivo. Asegúrese de que no se repitan. ¿Luego calcule el promedio de la población y la desviación estándar de esto 50 municipios?.

```python
# Seleccionar cincuenta municipios al azar sin repetición
municipios_aleatorios = dataset.sample(n=50, replace=False)

# Calcular el promedio de la población y la desviación estándar de los cincuenta municipios
promedio_poblacion = municipios_aleatorios['Poblacion'].mean()
desviacion_estandar_poblacion = municipios_aleatorios['Poblacion'].std()

print("El promedio de la población de los cincuenta municipios seleccionados al azar es:", promedio_poblacion)
print("La desviación estándar de la población de los cincuenta municipios seleccionados al azar es:", desviacion_estandar_poblacion)
```

### 8) Determine los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

```python
# Coordenadas de Madrid
coord_madrid = np.array([40.4165, -3.70256])  # Latitud y longitud de Madrid
# Calcular la distancia entre cada municipio y Madrid
dataset['Distancia_Madrid'] = np.sqrt((dataset['X'] - coord_madrid[0])**2 + (dataset['Y'] - coord_madrid[1])**2)

# Eliminar Madrid de los cálculos
data = dataset[dataset['Provincia'] != 'Madrid']

# Encontrar el municipio más cercano a Madrid
municipio_mas_cercano = dataset.loc[dataset['Distancia_Madrid'].idxmin()]

# Encontrar el municipio más lejano a Madrid
municipio_mas_lejano = dataset.loc[dataset['Distancia_Madrid'].idxmax()]

# Imprimir los resultados
print("El municipio más cercano a Madrid es:", municipio_mas_cercano['Municipio'], "en la provincia de", municipio_mas_cercano['Provincia'])
print("El municipio más lejano a Madrid es:", municipio_mas_lejano['Municipio'], "en la provincia de", municipio_mas_lejano['Provincia'])
```
